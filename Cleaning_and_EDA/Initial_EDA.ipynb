```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')
import seaborn as sns
```

```python
# import cleaned data
df = pd.read_csv('../Data/gss_subset_cleaned.csv')
```

```python
# check dtypes and nulls in current csv
df.info()
```

```python
# delete "Unnamed: 0" column
del df['Unnamed: 0']
```

```python
# change divorce to indicator variable (1 = has divorced / 0 = never divorced)
df['divorce'] = (df.divorce=='yes').astype(int)
```

```python
# list categorical columns and numerical columns for filtering
categorical_cols = ['marital', 'sex']
numerical_cols = df.columns.drop(categorical_cols)
```

```python
# look at histogram for each column (WARNING: this takes a while)
for col in df:
    df[col].value_counts().plot(kind='bar')
    plt.title(col+" distribution")
    plt.show()
```

```python
# look at happiness by age for male and female respondents
ax1 = plt.subplot(111)
df[df['sex']=='Male'].groupby(['age']).agg({'happy':
    np.mean}).plot(color='b', xlim = (0,90), ylim = (1.6,2),
    ax = ax1, legend = False)
df[df['sex']=='Female'].groupby(['age']).agg({'happy':
    np.mean}).plot(color='g',xlim = (0,90),
    ax = ax1)
plt.legend(['Male Avg Happiness', 'Female Avg Happiness'],
    bbox_to_anchor = (0.4,1))
plt.title('Happiness by age (lower is happier)')
plt.show()
```

```python
# plot happiness by marital status
ax2 = plt.subplot(111)
df.groupby('marital').agg({'happy': np.mean}).plot(kind='bar', legend=False, ax=ax2)
plt.title('Average Happiness by Marital Status')
plt.show()
```

```python
# average happiness by education level
ax3 = plt.subplot(111)
df.groupby('educ').agg({'happy': np.mean}).plot(kind='line', legend=False, ax=ax3)
plt.title('Average Happiness by Education Level')
plt.show()
```

```python
# self reported happiness over time
sns.factorplot(
    x = 'year',
    y = 'happy',
    data = df,
    hue = 'sex',
    size = 10
    )
```

```python
# happiness by marital status: violinplot
sns.factorplot(
    x = 'marital',
    y = 'happy',
    data = df,
    hue = 'sex',
    size = 10,
    kind='violin'
    )
```

```python
# look at happiness by number of children
sns.factorplot(
    x = 'childs',
    y = 'happy',
    data = df,
    hue = 'sex',
    size = 10,
    kind='bar'
    )
```

```python
# look at happiness by number of children < 6
sns.factorplot(
    x = 'babies',
    y = 'happy',
    data = df,
    hue = 'sex',
    size = 10,
    kind='bar'
    )
```

```python
# look at correlations in data
sns.heatmap(df.drop(categorical_cols, axis=1).corr())
```

```python
# look at happiness by political views (from 1: extremely liberal to 7: extremely conservative)
sns.factorplot(
    x = 'polviews',
    y = 'happy',
    data = df,
    size = 7,
    kind='bar'
    )
```

```python
# look at political views (from 1: extremely liberal to 7: extremely conservative) by age
sns.factorplot(
    x = 'age',
    y = 'polviews',
    data = df,
    size = 7,
    kind='point',
    hue = 'sex'
    )
```
